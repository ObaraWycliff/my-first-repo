```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/mac/Desktop/Africa Data School/December_Cohort_Resources/data/jumta.csv")
df
```

```python
#Create a line plot showing the monthly revenue over the two-year period
revenue =df.iloc[:,1]
month =df.iloc[:,0]
fig, ax = plt.subplots()
ax.plot(month, revenue)
plt.xticks(rotation=90)
plt.grid(color='black', linestyle='--', linewidth=0.2)
ax.set_title('Monthly revenue generation')
ax.set_xlabel('Months')
ax.set_ylabel('Revenue')
```

```python
#Create a bar plot showing the total number of orders per month
total_orders =df.iloc[:,2]
fig, ax = plt.subplots()
plt.bar(month, total_orders)
plt.xticks(rotation=90)
plt.title('Total sales per month ')
plt.xlabel('Months')
plt.ylabel('Total sales ')

```

```python
#Create a scatter plot showing the relationship between the number of orders and the average order value.
avg_order_value= df.iloc[:,-1]
plt.scatter(total_orders,avg_order_value)
plt.title('Relationship between orders and avg order-value ')
plt.xlabel('total orders')
plt.ylabel('avg_order_value')
```

```python
#Create a subplot containing all three visualizations created in steps 2-4 above .
# Create a subplot containing all three visualizations
avg_order_value = df.iloc[:, -1]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Scatter plot
axes[0, 0].scatter(total_orders, avg_order_value)
axes[0, 0].set_title('Relationship between orders and avg order-value')
axes[0, 0].set_xlabel('Total Orders')
axes[0, 0].set_ylabel('Average Order Value')

# Bar plot
axes[1, 0].bar(month, total_orders)
axes[1, 0].set_xticks(month)  # Set x-ticks to months
axes[1, 0].tick_params(axis='x', rotation=90)
axes[1, 0].set_title('Total sales per month')
axes[1, 0].set_xlabel('Months')
axes[1, 0].set_ylabel('Total Sales')

# Line plot
axes[1, 1].plot(month, revenue)
axes[1, 1].tick_params(axis='x', rotation=90)
axes[1, 1].grid(color='black', linestyle='--', linewidth=0.2)
axes[1, 1].set_title('Monthly revenue generation')
axes[1, 1].set_xlabel('Months')
axes[1, 1].set_ylabel('Revenue')

# Remove the empty subplot
fig.delaxes(axes[0, 1])

#Save as image
plt.savefig('subplot.png', dpi=300)

# Show the subplot
plt.show()


```
